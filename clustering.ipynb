```python
from sklearn.cluster import KMeans
import pandas as pd

attributes = pd.read_csv('G:\Courses\EE381V RL\project\installment.csv')
attributes_for_clustering = attributes[['性别',
                                        '借款金额',
                                        '第几次借款',
                                        '借款期限',
                                        '借款年利率',
                                        '院校城市与居住城市收入差距',
                                        '婚姻状况',
                                        '子女',
                                        '月收入',
                                        '住房条件',
                                        '最高学历',
                                        '借款动机',
                                        ]]
```

```python
cluster = KMeans(n_clusters=2, n_jobs=-1, random_state=0)
cluster.fit(attributes_for_clustering)
```

```python
labels = cluster.labels_
centers = cluster.cluster_centers_
```

```python
print('labels shape: ', labels.shape)
print('center of cluster1: ', centers[0], '\ncenter of cluster2: ', centers[1])
```

```python
attributes_new = attributes_for_clustering
attributes_new[['installment','user_id','loan_id']] = attributes[['第几笔还款','用户ID','款项ID']]
attributes_new['labels'] = labels
attributes_new = attributes_new.rename(columns={'性别': 'gender', 
                                               '借款金额': 'amount',
                                               '第几次借款': 'num_loan',
                                               '借款期限': 'duration',
                                               '借款年利率': 'year_ratio',
                                               '院校城市与居住城市收入差距': 'diff_city',
                                               '婚姻状况': 'marriage',
                                               '子女': 'kids',
                                               '月收入': 'month_in',
                                               '住房条件': 'housing',
                                               '最高学历': 'edu',
                                               '借款动机': 'motivation'})
attributes_new.head()
```

```python
trajectory = pd.read_csv('G://Courses/EE381V RL/project/trajectory.csv')
trajectory_new = trajectory.merge(attributes_new, how='left', on=['installment','user_id','loan_id'])
trajectory_new.head()
```
